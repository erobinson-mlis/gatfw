### Ford-Fulkerson Algorithm for Max Flow

```python
G = gr.Graph(4, undirected=False)
G.insert_edge(0,1,5.0)
G.insert_edge(0,2,10.0)
G.insert_edge(1,3,10.0)
G.insert_edge(2,1,7.0)
G.insert_edge(2,3,1.0)

l = G.make_edge_list()
nxel = []
for i in l:
    nxel.append((i.to_node, i.from_node, i.weight))
print(nxel)
```

```python
ff = max.ford_fulkerson(G, 0, 3)
print(f"Total flow: {ff.compute_total_flow()}")

ff.print_graph()
    
```
